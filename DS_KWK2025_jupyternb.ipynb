```python
import requests 
import pandas as pd
import config


#request datasets

#2010-2011 INDEX 0
response2010_2011 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2010-2011')
#2011-2012 INDEX 1
response2011_2012 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2011-2012')
#2012-2013 INDEX 2
response2012_2013 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2012-2013')
#2013-2014 INDEX 3
response2013_2014 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2013-2014')
#2014-2015 INDEX 4
response2014_2015 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2014-2015')
#2015-2016 INDEX 5
response2015_2016 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2015-2016')
#2017-2018 INDEX 6
response2017_2018 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2017-2018')
#2018-2019 INDEX 7
response2018_2019 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2018-2019')
#2019-2020 INDEX 8
response2019_2020 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2019-2020')
#2020-2021 INDEX 9 
response2020_2021 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2020-2021')
#2021-2022 INDEX 10
response2021_2022 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2021-2022')
#2022-2023 INDEX 11
response2022_2023 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2022-2023')
#2023-2024 INDEX 12
response2023_2024 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2023-2024')
#2024-2025 INDEX 13
response2024_2025 = requests.get(f'https://www.thesportsdb.com/api/v1/json/{config.api_key}/lookuptable.php?l=4328&s=2024-2025')

#access to API data
response_list = [response2010_2011, response2011_2012, response2012_2013, 
                 response2013_2014, response2014_2015, response2015_2016, 
                 response2017_2018, response2018_2019, response2019_2020, 
                 response2020_2021, response2021_2022, response2022_2023,
                 response2023_2024, response2024_2025 ]

```

```python
#create empty list to store JSON formats 
data_json = []

#convert all API calls in response_list to json format and then add to data_json 
for res in response_list: 
    transform_res = res.json()
    data_json.append(transform_res)
```

```python
#list used to delete columns in dataset
delete = ['idTeam', 'idStanding','strBadge', 'idLeague','strForm', 'strDescription', 'dateUpdated']

#new empty list of dataframes of all the data_json 
data_frames = []

index = 0
for index in range (len(data_json)-1):
    data_frames.append(pd.json_normalize(data_json[index]['table']).drop(delete, axis = 1))
    

```

```python
#example of before data cleaning

df_example = response2010_2011.json()
df_before_cleaning = pd.json_normalize(df_example['table'])
```
